```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
nifty = yf.download("^NSEI", period="5y")
print(nifty.head())
auto_adjust = True
```

```python
nifty.to_csv("nifty.csv")
print("Data of nifty saved successfully")
```

```python
print(nifty.shape)
```

```python
nifty["Returns"]=nifty["Close"].pct_change()
print(nifty[["Close","Returns"]].head())
```

```python
nifty["MA20"]=nifty["Close"].rolling(window=20).mean()
```

```python
nifty["MA50"]=nifty["Close"].rolling(window=50).mean()
```

```python
print(nifty[["Close","MA20","MA50"]].tail())
```

```python
plt.figure(figsize=(14, 7))
```

```python
plt.plot(nifty.index, nifty['Close'], label='Close Price', linewidth=2)
plt.plot(nifty.index, nifty['MA20'], label='20-day MA', alpha=0.7)
plt.plot(nifty.index, nifty['MA50'], label='50-day MA', alpha=0.7)
plt.title('Nifty 50: Price and Moving Averages (5 Years)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (₹)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
```

```python
plt.savefig('nifty50_chart.png', dpi=100)
plt.show()

print("✓ Chart saved  nifty50_chart.png")
```
